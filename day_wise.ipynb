```python
x_data = df['Deaths']
y_data = df['Recovered']
# y_data = df['']

        # Plotting
plt.scatter(x_data, y_data, label='Scatter Plot', color='red')
plt.xlabel('Deaths')
plt.ylabel('Recovered')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

```python
x_data = df['Deaths / 100 Cases']
y_data = df['Recovered / 100 Cases']
# y_data = df['']

        # Plotting
plt.scatter(x_data, y_data, label='Scatter Plot', color='red')
plt.xlabel('Deaths / 100 Cases')
plt.ylabel('Recovered / 100 Cases')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

```python
categories = ['1 week change', 'Deaths', 'Recovered',"Active"]
values = [25, 40, 30,50]

plt.bar(categories, values, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Graph Example')
plt.show()
```
